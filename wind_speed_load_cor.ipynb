```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
file_path = "./data/weatherkit_plus_load.csv"  # Update with actual path
df = pd.read_csv(file_path, parse_dates=['event_timestamp'])
df["wind_sin"] = np.sin(np.radians(df["weatherkit_forecast_wind_direction_deg"]))
df["wind_cos"] = np.cos(np.radians(df["weatherkit_forecast_wind_direction_deg"]))

# Compute correlation matrix including wind speed and load
correlation_matrix = df[["weatherkit_forecast_wind_speed_km_h", "wind_sin", "wind_cos", "load_MW"]].corr()

# Plot the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Between Wind Features and Load")
plt.show()
```
